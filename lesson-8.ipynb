```python

import numpy as np
import pandas as pd
 


import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
boston.keys()

```

```python
data = boston.data
```

```python
data.shape
```

```python
target = boston.target
```

```python
feature_names = boston.feature_names
feature_names
```

```python
x = pd.DataFrame(data, columns = feature_names)
x.head()
```

```python
y =  pd.DataFrame(target, columns = ["price"])
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=42)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_train_scaled = pd.DataFrame(scaler.fit_transform(x_train), columns=x_train.columns)
x_train_scaled
```

```python
x_test_scaled = scaler.fit_transform(x_test)
x_test_scaled = pd.DataFrame(x_test_scaled, columns = x_test.columns)
x_test_scaled
```

```python
from sklearn.manifold import TSNE
```

```python
y_train
```

```python
Z = TSNE(n_components=2, learning_rate=250, random_state=42).fit_transform(x_train_scaled)
```

```python
Z
```

```python

plt.plot(Z)
```

```python
plt.scatter(Z[:, 0], Z[:, 1])
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, max_iter=100, random_state=42)
```

```python
kmeans_preds = kmeans.fit_predict(x_train_scaled)
kmeans_preds.shape
```

```python
plt.scatter(Z[:, 0], Z[:, 1], c = kmeans_preds)
```

```python
y_train.shape
```

```python
print('Средний price')
print('Первый кластер: ', y_train[kmeans_preds == 0].mean())
print('Первый кластер: ', y_train[kmeans_preds == 1].mean())
print('Третий кластер: ', y_train[kmeans_preds == 2].mean())
```

```python
print('Средний CRIM')
print('Первый кластер: ', x_train.loc[kmeans_preds == 0, 'CRIM'].mean())
print('Первый кластер: ', x_train.loc[kmeans_preds== 1, 'CRIM'].mean())
print('Третий кластер: ', x_train.loc[kmeans_preds == 2, 'CRIM'].mean())
      
```

```python
kmeans_pred_test = kmeans.predict(x_test_scaled)
print(y_test.shape)
print(x_test.shape)
kmeans_pred_test.shape

```

```python
print('Средний price')
print('Первый кластер: ', y_test[kmeans_pred_test == 0].mean())
print('Первый кластер: ', y_test[kmeans_pred_test == 1].mean())
print('Третий кластер: ', y_test[kmeans_pred_test == 2].mean())
```

```python
print('Средний CRIM:')
print('Первый кластер: ', x_test.loc[kmeans_pred_test == 0, 'CRIM'].mean())
print('Первый кластер: ', x_test.loc[kmeans_pred_test== 1, 'CRIM'].mean())
print('Третий кластер: ', x_test.loc[kmeans_pred_test == 2, 'CRIM'].mean())
```
